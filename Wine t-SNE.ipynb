```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r"run3.csv", encoding="cp1252", index_col=0)
data["wine_id"] = data.index
```

```python
for i in [5.0,10.0,20.0,30.0,50.0,60.0,100.0]:
    print(i)
    norminvdf = np.load("norminvdf")
    pca = PCA(n_components=30)
    norminvdf_pca = pca.fit_transform(norminvdf)
    tSNE = TSNE(n_components=2, random_state=0, perplexity=i)
    norminvdftSNE = tSNE.fit_transform(norminvdf_pca)
    plt.scatter(norminvdftSNE[:,0],norminvdftSNE[:,1])
    plt.show()
    cols = ["red", "white", "yellow", "pink", "brown", "cyan", "black"]
    for i, c in enumerate(data.Colour.unique()):
        plt.scatter(norminvdftSNE[(data.Colour==c).values,0],
                    norminvdftSNE[(data.Colour==c).values,1], 
                    color=cols[i],
                    edgecolor='black')
    plt.show()
```
